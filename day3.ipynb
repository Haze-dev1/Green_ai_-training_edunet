If your friend opens the phone, the phone only gives them access to some things not all 

```python
# All phone features
all_features = {"photos", "contacts", "messages", "banking", "settings", "camera", "music"}

# Features available to friend (limited access)
friend_access = {"camera", "music"}

print("Your friend can access:", friend_access)
print("Your friend cannot access:", all_features - friend_access)
```
