```python
account1 = {
    "name": "Devansh",
    "age": 20,
    "city": "Nagpur",
    "expense": 5000
}

account2 = {
    "name": "Rahul",
    "age": 25,
    "city": "Mumbai",
    "expense": 7000
}

total_expense = account1["expense"] + account2["expense"]
print("Total expense:", total_expense)
total_balance = 15000 - total_expense
print("Total balance:", total_balance)
```

```python
account1 = {"balance": 5000}
transaction_limit = 2
transactions = 0
for i in range(5):
    if transactions >= transaction_limit:
        print("Daily transaction limit is reached.")
        break
    if account1["balance"] >= 1000:
        account1["balance"] -= 1000
        transactions += 1
        print(f"Transaction {transactions}: Paid 1000. Remaining balance: {account1['balance']}")
    else:
        print("Insufficient balance.")
        break
```
