```python
import tensorflow as tf
from tensorflow.keras.datasets import california_housing
cal_hdf = california_housing.load_data()
type(cal_hdf)
```

```python
len(cal_hdf)
```

```python
(x_train, y_train), (x_test, y_test) = cal_hdf
```

```python
x_train.shape
```

```python
x_train[:1]
```

```python
y_train[:1]
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
#scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
x_train_scaled[:1]
```

```python
#creating neural network architecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(128, activation= "relu")) 
model.add(Dense(1)) #Output layer
```

```python
#Compiling the NN
model.compile(loss = 'mean_absolute_error', optimizer = 'sgd', metrics = ['mae'])
```

```python
model.summary()
```

```python
#Training the NN model
history = model.fit(x_train_scaled,y_train, epochs = 60, validation_split = 0.2)
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label = 'train_loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.evaluate(x_test_scaled, y_test)
```

```python
x_new = x_test_scaled[:3]
predictions = model.predict(x_new)
predictions
```

```python
model.save('california_housing_model.keras')
```

```python
from tensorflow.keras.datasets import mnist
mnist_data = mnist.load_data()
type(mnist_data)
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist_data
```

```python
#reshape and normalize the array
train_images_scaled = train_images.reshape((60000, 28 * 28))/255.0
test_images_scaled = test_images.reshape((10000, 28 * 28))/255.0
```

```python
train_images_scaled[0]
```

```python
#creating neural network architecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
digit_nn_model = Sequential()
digit_nn_model.add(Dense(256, activation= "relu")) 
digit_nn_model.add(Dense(128, activation="relu"))
digit_nn_model.add(Dense(10, activation="softmax")) 
```

```python
digit_nn_model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
#train the model 
history = digit_nn_model.fit(train_images_scaled, train_labels, epochs = 40, batch_size= 64 ,validation_split = 0.2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label = 'train_loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label = 'train_loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.plot(history.history['accuracy'], label = 'train_accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#creating neural network architecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
digit_nn_model = Sequential()
digit_nn_model.add(Dense(256, activation= "relu"))
digit_nn_model.add(Dropout(0.2)) 
digit_nn_model.add(Dense(128, activation="relu"))
digit_nn_model.add(Dropout(0.2))
digit_nn_model.add(Dense(10, activation="softmax")) 
```

```python
digit_nn_model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
#train the model 
history = digit_nn_model.fit(train_images_scaled, train_labels, epochs = 40, batch_size= 64 ,validation_split = 0.2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label = 'train_loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label = 'train_loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.plot(history.history['accuracy'], label = 'train_accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#evaluate on the test data
loss, acc = digit_nn_model.evaluate(test_images_scaled, test_labels)
print(f"Accuract of the og model is:{acc*100}")

```
